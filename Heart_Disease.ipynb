```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
#Loading the data to pandas dataframe

heart_data = pd.read_csv('/content/heart_disease.csv')
```

```python
# Checking the distribution of Target variable

heart_data['condition'].value_counts()
```

1 --> Defective Heart

0 --> Healthy Heart

Splitting the features and Target columns

```python
X = heart_data.drop(columns='condition', axis=1)
Y = heart_data['condition']
```

Splitting data into training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training with Logistic Regression Model

```python
model = LogisticRegression()
```

```python
# Training the model with trainng data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data.
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print('Accuracy on Training data: ', training_data_accuracy)
```

```python
# accuracy on testing data.
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print('Accuracy on Test data: ', test_data_accuracy)
```

Building a Predictive system

```python
input_data = (61,1,0,134,234,0,0,145,0,2.6,1,2,0)

# change input data to numpy array
numpy_input_data = np.asarray(input_data)
print(numpy_input_data.shape)

# reshape the numpy array since we are predicting for only one instance
reshaped_input_data = numpy_input_data.reshape(1, -1)
print(reshaped_input_data.shape)

prediction = model.predict(reshaped_input_data)
print(prediction)

if prediction[0] == 0:
  print('Patient does not have Heart Disease')
else:
  print('Patient has Heart Disease')
```
